```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("hw.txt", header = None)
```

```python
data = np.array(data)
```

```python
new_data = data.reshape(len(data)//2, 2)
```

```python
print(new_data)
```

```python
df = pd.DataFrame(new_data, columns = ["Height(cm)", "Weight(lb)"])
```

```python
df
```

```python
df["Height(m)"] = df["Height(cm)"]/100
df["Weight(kg)"] = df["Weight(lb)"]/2.2
```

```python
df
```

```python
df.index = np.arange(1, len(df) + 1)
```

```python
df.drop(columns=["Height(cm)", "Weight(lb)"], inplace = True)
```

```python
for i in range(0, len(df.index)):
    df["Person"] = df.index
```

```python
df
```

```python
df["BMI"] = df["Weight(kg)"]/df["Height(m)"]**2
```

```python
df
```

```python
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
cut_bins = [0, 18.5, 24.9, 29.9, df["BMI"].max()]
df['Category'] = pd.cut(df['BMI'], bins=cut_bins, labels=bmi_labels)
```

```python
df
```

```python
grouped = df.groupby('Category')

overall_group = []
for category, group in grouped:
    print(f"{category}:\n")
    for index, row in group.iterrows():
        print(f"Person-{row['Person']}: {row['BMI']:.4f}")
    group_size = len(group)
    average_bmi = group['BMI'].mean()
    print(f"Group Size: {group_size}")
    print(f"Average BMI for this group is: {average_bmi:.4f}\n")
    overall_group.extend(group['BMI'])

overall_group_size = len(overall_group)
overall_average_bmi = sum(overall_group) / overall_group_size
print(f"Overall Group Size: {overall_group_size}")
print(f"Overall Average BMI: {overall_average_bmi:.4f}")
```
